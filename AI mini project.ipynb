```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import traceback

# --------------------------
# 1) LOAD/DEFINE your dataframe
# If you already have a dataframe named `df`, you can skip the demo loader.
# Otherwise replace the read_csv path with your file path:
# df = pd.read_csv('path/to/your/Telco-Customer-Churn.csv')

try:
    df  # check if df exists
except NameError:
    print("Variable 'df' not found — creating a tiny demo dataframe for testing.")
    df = pd.DataFrame({
        'PhoneService': ['Yes','No','Yes','Yes','No'],
        'MultipleLines': ['No','No phone service','Yes','No','Yes'],
        'InternetService': ['DSL','Fiber optic','No','DSL','Fiber optic'],
        'OnlineSecurity': ['No','Yes','No','Yes','No'],
        'OnlineBackup': ['Yes','No','No','Yes','Yes'],
        'DeviceProtection': ['No','Yes','No','Yes','No'],
        'TechSupport': ['No','No','Yes','No','Yes'],
        'StreamingTV': ['No','Yes','No','Yes','No'],
        'StreamingMovies': ['No','Yes','No','Yes','No'],
        'Churn': ['No','Yes','No','No','Yes']
    })

# --------------------------
# 2) Columns to plot
cols = [
    'PhoneService','MultipleLines','InternetService','OnlineSecurity',
    'OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies'
]

# --------------------------
# 3) Clean column names (strip spaces and stray quotes)
df = df.copy()
df.columns = df.columns.astype(str).str.strip()
# remove stray single/double quotes inside column names if present
df.columns = df.columns.str.replace(r"['\"]", "", regex=True)

# Check for missing columns
missing = [c for c in cols if c not in df.columns]
if missing:
    print("ERROR: these columns are missing from the dataframe:", missing)
    print("Available columns (first 20):", list(df.columns)[:20])
    # Optionally attempt fuzzy match or show suggestions, but stop here so you can fix names
    raise SystemExit("Fix column names or update the 'cols' list and re-run.")

# --------------------------
# 4) Plotting function (robust)
def plot_countplots(df, cols, hue=None, n_cols=3, figsize=(15,12), na_label='Missing'):
    n_rows = (len(cols) + n_cols - 1) // n_cols
    plt.figure(figsize=figsize)
    for i, col in enumerate(cols, 1):
        try:
            ax = plt.subplot(n_rows, n_cols, i)
            # fill NA so countplot doesn't choke on nan
            s = df[col].fillna(na_label).astype(str)
            # keep original df for hue but ensure hue column cleaned/exists
            order = s.value_counts().index
            if hue and hue in df.columns:
                sns.countplot(data=df.assign(_tmp_col=s), x='_tmp_col', hue=hue, order=order)
                plt.xlabel(col)  # set readable x-label
            else:
                sns.countplot(data=df.assign(_tmp_col=s), x='_tmp_col', order=order)
                plt.xlabel(col)
            plt.title(col)
            plt.xticks(rotation=30, ha='right')
            plt.ylabel('Count')
        except Exception as e:
            print(f"Failed to plot '{col}': {e}")
            traceback.print_exc()
    plt.tight_layout()
    plt.show()

# --------------------------
# 5) Call the function
# If you have a churn column named exactly 'Churn' and want split bars by churn, keep hue='Churn'
# Otherwise call with hue=None
plot_countplots(df, cols, hue='Churn')
```
